### B. Data Preparation

#### Detecting Missing Values

```python
dataNBA17.isnull().sum()
```

```python
# Drop column 'Unnamed:0 & column with missing values for entire rows'

dataNBA17.drop(['Unnamed: 0', 'blanl', 'blank2'], axis=1, inplace=True)
```

```python
# Figure out the distribution of variables that have missing values

plt.figure(figsize=(15, 7))

plt.subplot(3,3,1)
dataNBA17['TS%'].plot.hist(bins=12, alpha=0.5)

plt.subplot(3,3,2)
dataNBA17['3PAr'].plot.hist(bins=12, alpha=0.5)

plt.subplot(3,3,3)
dataNBA17['FTr'].plot.hist(bins=12, alpha=0.5)

plt.subplot(3,3,4)
dataNBA17['TOV%'].plot.hist(bins=12, alpha=0.5)

plt.subplot(3,3,5)
dataNBA17['FG%'].plot.hist(bins=12, alpha=0.5)

plt.subplot(3,3,6)
dataNBA17['3P%'].plot.hist(bins=12, alpha=0.5)

plt.subplot(3,3,7)
dataNBA17['2P%'].plot.hist(bins=12, alpha=0.5)

plt.subplot(3,3,8)
dataNBA17['eFG%'].plot.hist(bins=12, alpha=0.5)

plt.subplot(3,3,9)
dataNBA17['FT%'].plot.hist(bins=12, alpha=0.5)
```

```python
# Imputing Missing Values for numerical data type, using mean for data with normal distribution and using median for data with skewness

dataNBA17['TS%'] = dataNBA17['TS%'].fillna(dataNBA17['TS%'].median())
dataNBA17['3PAr'] = dataNBA17['3PAr'].fillna(dataNBA17['3PAr'].mean())
dataNBA17['FTr'] = dataNBA17['FTr'].fillna(dataNBA17['FTr'].median())
dataNBA17['TOV%'] = dataNBA17['TOV%'].fillna(dataNBA17['TOV%'].median())
dataNBA17['FG%'] = dataNBA17['FG%'].fillna(dataNBA17['FG%'].mean())
dataNBA17['3P%'] = dataNBA17['3P%'].fillna(dataNBA17['3P%'].median())
dataNBA17['2P%'] = dataNBA17['2P%'].fillna(dataNBA17['2P%'].mean())
dataNBA17['eFG%'] = dataNBA17['eFG%'].fillna(dataNBA17['eFG%'].mean())
dataNBA17['FT%'] = dataNBA17['FT%'].fillna(dataNBA17['FT%'].median())
```

#### Data Cleaning

```python
# Change the data type variable 'Year' from float64 to integer

dataNBA17 = dataNBA17.astype({"Year": int})
```

```python
# Remove duplicates player names while keeping the last row for each names, because the last one were the last teams for each player (based on Googling)

dataNBA17 = dataNBA17.drop_duplicates(subset=['Player'], keep='last')
```

```python
# Change the index start from one 

dataNBA17.index = np.arange(1, len(dataNBA17)+1)
```

### C. Answering Question

#### [1] Who is the youngest and oldest player in the NBA in 2017 for each team (Tm) ?

```python
# The youngest player in each Team 
datayoungest = dataNBA17.groupby(['Tm', 'Player']).agg({'Age':np.min}).reset_index() \
          .sort_values(['Tm', 'Age'], ascending=[1, 1]).groupby('Tm').head(1)
datayoungest
```

```python
# The oldest player in each Team 
dataoldest = dataNBA17.groupby(['Tm', 'Player']).agg({'Age':np.max}).reset_index() \
          .sort_values(['Tm', 'Age'], ascending=[1, 0]).groupby('Tm').head(1)
dataoldest
```

#### [2] Which player has the most minutes played (MP) in each position (Pos)?

```python
# The most minutes played (MP) in each position (Pos)
datamostMP = dataNBA17.groupby(['Pos', 'Player']).agg({'MP':np.max}).reset_index() \
          .sort_values(['Pos', 'MP'], ascending=[1, 0]).groupby('Pos').head(1)
datamostMP
```

#### [3] Which team has the highest average total rebound percentage (TRB%), assist percentage (AST%), steal percentage (STL%), and block percentage (BLK%)?

```python
datahighestAvg = dataNBA17.groupby(['Tm']).agg({'MP':np.average, 'AST%':np.average, 'STL%': np.average, 'BLK%': np.average}).reset_index() \
          .sort_values(['MP', 'AST%', 'STL%', 'BLK%'], ascending=[0, 0, 0, 0]).head(1)
datahighestAvg
```

#### [4] Who is the best player in your opinion based on his record stats? note: you can refer to variables point (PTS), assists, rebounds, or anything else. A combination of several variables would be nice.

```python
# The best player based on the highest PTS, AST, and eFG%

thebestPlayer = dataNBA17.groupby(['Player']).agg({'PTS':np.max, 'AST':np.max, 'eFG%': np.max}).reset_index() \
          .sort_values(['PTS', 'AST', 'eFG%'], ascending=[0, 0, 0]).head(1)
thebestPlayer 
```

#### [5]	Which team has the best average stat record of their players? Note: you can refer to points, assists, rebounds, or anything else. A combination of several variables would be nice

```python
# The best team based on the highest PTS, AST, and eFG%

thebestTeam = dataNBA17.groupby(['Tm']).agg({'PTS':np.max, 'AST':np.max, 'eFG%': np.max}).reset_index() \
          .sort_values(['PTS', 'AST', 'eFG%'], ascending=[0, 0, 0]).head(1)
thebestTeam 
```
