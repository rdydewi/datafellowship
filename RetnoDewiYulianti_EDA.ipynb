```python
# Import function that will use for this analysis

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Importing Dataset

epl1819 = pd.read_csv("epl_1819.csv")
epl1819.head()
```

```python
# Statistical Descriptive

epl1819.describe()
```

### 1. Is this data Clean?

```python
epl1819.info()
```

##### There is no missing value, but there are some data type that must be change from categorical into numerical values, and also renaming the variable to not space name.

```python
epl1819["attack_passes"] = [float(str(i).replace(",", "")) for i in epl1819["attack_passes"]]
epl1819["attack_passes_long"] = [float(str(i).replace(",", "")) for i in epl1819["attack_passes_long"]]
epl1819["attack_passes_back"] = [float(str(i).replace(",", "")) for i in epl1819["attack_passes_back"]]
epl1819["defence_clearances"] = [float(str(i).replace(",", "")) for i in epl1819["defence_clearances"]]
```

```python
# call rename () method
epl1819.rename(columns={'finance _live_games_televised': 'finance_live_games_televised', 'finance _tv_revenue': 'finance_tv_revenue', 'finance _team_market': 'finance_team_market', 'finance _market_average': 'finance_market_average'}, inplace=True)
```

```python
epl1819.info()
```

```python
# change index number start from 1

epl1819.index = np.arange(1, len(epl1819)+1)
```

```python
Q1 = epl1819.quantile(0.25)
Q3 = epl1819.quantile(0.75)
IQR = Q3 - Q1
print(IQR)
```

### Identifying Outliers

```python
print(epl1819 < (Q1 - 1.5 * IQR)) |(epl1819 > (Q3 + 1.5 * IQR))
```

### 3.	Which team has the best attack?  

Source: https://www.premierleague.com/stats/top/clubs/ 

```python
# Adding new column of total attack

epl1819['total_attack'] = epl1819['attack_shots'] + epl1819['attack_shots_on_target'] + epl1819['attack_goals_headed'] + epl1819['attack_goals_penalty'] + epl1819['attack_goals_box'] + epl1819['attack_goals_outsidebox'] + epl1819['attack_goals_counter'] + epl1819['attack_goals_freekick']
```

```python
best_team_attack = epl1819[['Team', 'total_attack']].sort_values(by='total_attack', ascending=False)
best_team_attack.index = np.arange(1, len(best_team_attack)+1)
best_team_attack.head(5)
```

### 4.	Which team has the best defence? 

Source: https://www.premierleague.com/stats/top/clubs/ 

```python
# Adding new column of total deffence

epl1819['total_defence'] = epl1819['defence_saves'] + epl1819['defence_blocks'] + epl1819['defence_interceptions'] + epl1819['defence_tackles'] + epl1819['defence_tackles_last_man'] + epl1819['defence_clearances'] + epl1819['defence_clearances_headed'] + epl1819['defence_penalty_conceeded']
```

```python
best_team_defence = epl1819[['Team', 'total_defence']].sort_values(by='total_defence', ascending=True)
best_team_defence.index = np.arange(1, len(best_team_defence)+1)
best_team_defence = best_team_defence.astype({"total_defence": int})
best_team_defence.head(5)
```

### 5.	Which team is good in the financial aspect? 

```python
best_tv_revenue = epl1819[['Team', 'finance_tv_revenue']].sort_values(by='finance_tv_revenue', ascending=False).head(5)
best_tv_revenue.index = np.arange(1, len(best_tv_revenue)+1)
best_tv_revenue
```

```python
plt.figure(figsize=(7, 7))
values = best_tv_revenue["finance_tv_revenue"]
clrs = ['cadetblue' if (x < max(values)) else 'lawngreen' for x in values ]
sns.barplot(x=best_tv_revenue["finance_tv_revenue"], y= best_tv_revenue["Team"], palette = clrs)
plt.show()
```

```python
best_team_market = epl1819[['Team', 'finance_team_market']].sort_values(by='finance_team_market', ascending=False).head(5)
best_team_market.index = np.arange(1, len(best_team_market)+1)
best_team_market
```

```python
best_market_avg = epl1819[['Team', 'finance_market_average']].sort_values(by='finance_market_average', ascending=False).head(5)
best_market_avg.index = np.arange(1, len(best_market_avg)+1)
best_market_avg
```

```python
plt.figure(figsize=(15, 15))
fig, axs = plt.subplots(1, 2, sharex=True, figsize=(10,5))
fig.suptitle('MARKET VALUE IN ALL CLUBS IN PREMIER LEAGUE 1819')
axs[0].set_title('Top 5 Finance Team Market')
axs[1].set_title('Top 5 Finance Market Average')


values1 = best_team_market["finance_team_market"]
clrs1 = ['slategrey' if (x < max(values1)) else 'mediumvioletred' for x in values1 ]
sns.barplot(x=best_team_market["Team"], y= best_team_market["finance_team_market"], palette = clrs1, ax=axs[0])

values2 = best_market_avg["finance_market_average"]
clrs2 = ['slategrey' if (x < max(values2)) else 'orangered' for x in values2 ]
sns.barplot(x=best_market_avg["Team"], y= best_market_avg["finance_market_average"], palette = clrs2, ax=axs[1])

plt.show()
```
